```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df1 = pd.read_csv('spam.csv', encoding='latin1')
df1
```

```python
df1.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
df1
```

```python
df1['v2'] = df1['v2'].str.lower()
df1
```

```python
import re
df1['v2'] = df1['v2'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
df1['v2'] = df1['v2'].apply(lambda x: re.sub(r'\d+', '', x))
df1
```

```python
df1['v1'] = df1.v1.map({'spam':0, 'ham':1})
df1
```

```python
X = df1['v2'].values
y = df1['v1'].values
y = y.astype('int')
```
